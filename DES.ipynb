```python
def XOR_function(a , b):

    data = np.zeros(len(a))

    for i in range(len(a)):
        data[i] = int(a[i]) ^ int(b[i])
        

    return data
```

```python
def table_np(table , array):
    new_array = np.array([])
    for i in range(len(table)):
        #we will take the value of the position i and put it in the new array 
        new_array = np.insert(new_array , i , int(array[table[i]-1]))

    return new_array

#team leader valeo

# def table(table , array):
#     new_array = []
#     for i in range(len(table)):
#         new_array.insert(i ,int(array[table[i]-1]))

#     return new_array
```

```python
def right_shift(bits , number_of_shifts):
    #we take jsut the end bits we want to move forward + the same string - the final bits 
    #by other word we take the string from index 3 but from the end not the beganing + the string from the beganing to index 3 from the end
    return bits[-number_of_shifts:] + bits[:-number_of_shifts]

# -----Test----
arr = "1234"
new = right_shift(arr , 2)
print(new)
    
```

```python
def left_shift(bits , number_of_shifts):
    #we take all the data except the first bits we want to move backward + the first bits 
    #by other word we take the string from index n instead of index 0 + the string from 0 index to n
    return bits[number_of_shifts:] + bits[:number_of_shifts]
```

```python
def S_Box(s , table):
    
    #take the first and final bits to make the index of the row 
    row = np.concatenate((s[0:1], s[5:6]))

    #change the type of the row to string, so we can merge the bits
    #Note: we didnt change it from the beginig because the concatenate function work with intgers only
    row = ''.join(row.astype(str))
    
    #then we change the binary number to decimal, so that we can search with the intger index in the table 
    decimal_row = int(row, 2)

    #we did the same steps with the column, we just take the 4 mid bites to get the number of the column
    column = s[1:5]
    column = ''.join(column.astype(str))
   
    decimal_column = int(column, 2)

    #we get the value from the box 
    s_res = table[decimal_row][decimal_column]

    #change it to binary. we removed this (note that we put this [2:].zfill(4) to print it in 4 bits)
    s_res_bin = bin(s_res)

    #take the final 4 bits from the result (the old result 0b1110 --> new result 1110)
    s_res_bin = s_res_bin[2:6]

    if len(s_res_bin) < 4:
        for i in range(4 - len(s_res_bin)):
            s_res_bin = '0' + s_res_bin

    return s_res_bin
```

```python
def f_function(right , key):

    #first: we do the expansion to the right to change its size from 32 bits to 48 bits 
    Expantion_Table = np.array([32,1,2,3,4,5,
                                4,5,6,7,8,9,
                                8,9,10,11,12,13,
                                12,13,14,15,16,17,
                                16,17,18,19,20,21,
                                20,21,22,23,24,25,
                                24,25,26,27,28,29,
                                28,29,30,31,32,1])

    right_Ex = table_np(Expantion_Table , right)
    # print("riht(E): " , right_Ex)
    #second: we do XORing operation to the new right with the key
    right_xor = XOR_function(right_Ex , key)
    #right_xor = right_xor.astype(int)
    # print("riht(XOR): " , right_xor)
    #third: we take it as a input to the s-box to do some operations
    right_Sbox = S_Box_combination(right_xor)
    #print(right_xor.dtype)
    # print("right(s-box): " , right_Sbox)
    p_table = np.array([16, 7, 20, 21, 29, 12, 28, 17,
                        1, 15, 23, 26, 5, 18, 31, 10,
                        2, 8, 24, 14, 32, 27, 3, 9,
                        19, 13, 30, 6, 22, 11, 4, 25])

    #finally we change the positions of the bits using the p table, then the new right is ready!
    right_p = table_np(p_table , right_Sbox)
    # print("riht(p): " , right_p)
    return right_p
```
